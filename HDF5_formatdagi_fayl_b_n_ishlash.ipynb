# HDF5 formatni avzalligi boshqa dasturlash tiilarida ham buni ochish mumkin.
# Ilmiy izlanishlar qilganda katta fayllarni saqlash uchun ishlatiladigan formatlardan biri.
# Katta jadvallarni saqlganda siqilgan holatda saqlaydi

```python
# bu format bilan ishlash uchun yangi obj yaratib olamiz
hdfobj=pd.HDFStore("fayl_name.h5") # biz hdfobj o'zgaruvchiga nimani saqlasak fayl_name degan faylgaham saqlanadi
# biz manashu o'zgaruvchiga ber nechta faylni saqlasak bo'ladi
hdfobj.put("avtodata",df) # "avtodata" bu kalit hisoblanadi
# endi uzbek degan o'zgaruvchini yuklaymiz
hdfobj.put("uzdata",uzbek)
```

```python
# biz bu hdfobj ni kalitlarini ko'rishimiz mumkin
hdfobj.keys()
```

```python
# biz saqlagan fayllarimizga huddi ro'yhatga murojat qilganimiz kabi      hdfobj["uzdata"]    deb chaqirib olsak bo'ladi
hdfobj['uzdata'] # ko'rishimiz mumkin bizga jadvalni chiqarib berdi
```

```python
hdfobj["avtodata"]
```

```python
# biz agar fayl_name.h5 bilan ishlab bo'lgan bo'lsak faylni 
# yopishimiz kerak va fayl yopilgandan keyin undan obj chaqirib bo'lmaydi
hdfobj.close()
```

```python
hdfobj['uzdata'] # ko'rishingiz mumkin hato berdi/
```

```python
# agar faylni ochmoqchi bo'lsak huddi boshidagidek yozamiz
hdfobj=pd.HDFStore("fayl_name.h5")
```

```python
hdfobj["uzdata"] # ko'rishingiz mumkin faylni ochdi
```

```python
hdfobj.keys()
```

```python
# hdfobj yopiq turgan holidaham faylni chaqirib olishimiz mumkin
hdfobj.close() 
```
