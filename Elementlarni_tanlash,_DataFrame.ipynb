# TAKROROLASH

```python
uz={2016:28.0,2018:28.9,2020:32.0,2022:33.0}
tj={2016:25.2,2018:26.9,2020:27.4,2022:30.0}
kz={2016:38.0,2018:38.9,2020:40.0,2022:42.0}
af={2016:16.0,2018:16.9,2020:17.5,2022:19.0}
tur={2016:28.0,2018:28.9,2020:32.0,2022:33.0}
dic_data={"Uzbekistan":uz,
          "Tajikistan":tj,
          "Kazahstan":kz,
          "Afganistan":af,
          "Turkmenistan":tur}

df5=pd.DataFrame(dic_data)
```

```python
df5
```

```python
# agar data bizga Series formatida kerak bo'lsa  quyidagicha yozamiz
df5["Uzbekistan"]
```

```python
# indexlarni o'zgartirishimiz mumkin
df5.index=[16,18,20,22]
```

```python
df5
```

```python
# agar DF formatida kerak bo'lsa  quyidagicha yozamiz
df5[["Uzbekistan"]]
```

```python
# agar bizga 30 mln.dan  ko'p aholi bor yil kerak bo'lsa quyidagicha yozamiz
df5[["Uzbekistan"]][df5["Uzbekistan"]>30] 
```

```python
df5
```

```python
# df5ni o'qlarini almashtiramiz
df6=df5.transpose(copy=True)
```

```python
df6
```

```python
# ko'rishingiz mumkin jadval o'zgardi
df6[[20]][df6[20]>30] # 20 yilda aholi soni 30 mln dan ko'p bo'lgan davlatni ajratib olamiz
```

```python
# df6 dan 30 mlndan ko'p aholini tekshirish uchun quyidagicha yozamiz
df6>30 # 30 mln dan ko'p aholini True dedi
```
